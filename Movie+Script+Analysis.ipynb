# Analyzing Leonardo Dicaprio's "The Revenant" movie script

The [Internet Movie Script Database](http://www.imsdb.com/) contains the text of screenplays for thousands of movies, including many current releases. .

```python
import requests
from bs4 import BeautifulSoup, NavigableString, Tag
import nltk
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
screenplay_raw = requests.get('http://www.imsdb.com/scripts/Revenant,-The.html').text
```

Turning the `screenplay_raw` into soup.

```python
screenplay_soup = BeautifulSoup(screenplay_raw,'lxml')
```

Here I used google chrome's JavaScript console to identify the part of the script's HTML tree containing the body of the script. Then I used BeautifulSoup's [`.find_all()`](https://www.crummy.com/software/BeautifulSoup/bs4/doc/#find-all) method to extract this information. **(Hint: it's beneath the `<td class="scrtext">` tag.)**

```python
screenplay_body = screenplay_soup.find_all("pre")
```

Many (but not guaranteed to be all) scripts appear to use the `<b>` tags to mark when a character speaks. Useing the `.find_all()` method on the `screenplay_body` I captured all of these headings and saved them as `screenplay_headings`.

```python

screenplay_headings = screenplay_body[0].find_all("b") 
len(screenplay_headings)
```

What's are examples of screenplay headings?

```python

screenplay_headings[90:100]
```

It looks like there's lots of white space in these headings. Using Python's string methods like [`.strip()`](https://docs.python.org/3.5/library/stdtypes.html#string-methods) I looped through these headings and cleaned them up to store as `cleaned_headings`. 

```python
cleaned_headings = list()

for heading in screenplay_headings:

    cleaned_heading = heading.text.strip()
    cleaned_headings.append(cleaned_heading)    
   
```

```python
type(screenplay_headings[90])
```

What do these headings look like now?

```python
cleaned_headings[1:895]
```

Made a dictionary that is keyed by the heading and has a value for the number of times the heading occurs. 


```python
heading_counter = dict()
for text in cleaned_headings:
    if text not in heading_counter:
        heading_counter[text] = 1
    else:
        heading_counter[text] += 1
       
            
```

Using my `sorting_dict` function to give the top 20 headings.

```python
from operator import itemgetter

def sorting_dict(d):
    _sorted = sorted(d.items(),key=itemgetter(1),reverse=True)
    return _sorted[:895]

sorting_dict(heading_counter)
```

## Problem 3: Normalizing screenplay text



```python
porter = nltk.PorterStemmer()
wnl = nltk.WordNetLemmatizer()
```

I created a `line_tokens` list containing all the lower-case tokens by looping over each line of the `cleaned_lines` and using the `nltk.word_tokenize` function.

```python
line_tokens = list()

for line in cleaned_lines:
    lowered = line.lower()
    tokens = nltk.word_tokenize(lowered)
    for token in tokens:
        line_tokens.append(token)
line_tokens[150:200]        
```

NLTK will tell you if each word is alphabetic or a punctuation phrase.

```python
for token in line_tokens[:20]:
    print(token.rjust(10),token.isalpha())
```

NLTK also has a list of common English stopwords.

```python
stopwords = nltk.corpus.stopwords.words('english')
stopwords
```

Created a `cleaned_tokens` list that removes the stopwords and the punctuation.

```python
clean_tokens = list()
for token in line_tokens:
    if token not in stopwords and token.isalpha():
        clean_tokens.append(token)
        
clean_tokens[:20]   

```

Now use the `porter` PorterStemmer to stem each token.

```python
porter_stemmed_tokens = list()

for token in clean_tokens:
    stemmed_tokens = porter.stem(token)
    porter_stemmed_tokens.append(stemmed_tokens)

porter_stemmed_tokens    
```

Used the `wnl` WordNet lemmatizer to lemmatize each token as a comparison.

```python
wordnet_lemmatized_tokens = list()
for token in clean_tokens:
    lemmatized_token = wnl.lemmatize(token.lower())
    wordnet_lemmatized_tokens.append(lemmatized_token)
    
wordnet_lemmatized_tokens    
```

Created a counter dictionary for each bag of tokens (stemmed and lemmatized).

```python
#I could not figure out why I'm getting this error 
#I know this is the right method I just can't get it to run


porter_freq = dict()
for text in porter_stemmed_tokens:
    if text not in porter_freq:
        porter_freq[text] = 1
    else:
        porter_freq[text] += 1
porter_freq

lemma_freq = dict()
for text in wordnet_lemmatized_tokens:
    if text not in lemma_freq:
        lemma_freq[text] = 1
    else:
        lemma_freq[text] += 1
lemma_freq        
        

        
        
        
        
```
